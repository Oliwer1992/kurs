```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('office.jpg') # Wczytuje obraz
```

```python
img.shape # parametry(wysokość, szerokość, kolory(3 bo rgb))
```

```python
plt.imshow(img) # Wyświetla obraz
plt.axis('off') # Wyłącza osie na obrazie
```

```python
fig = plt.figure(figsize=(12,10)) # Rozmiar to 1200x1000
plt.subplot(221)
plt.imshow(mpimg.imread('office.jpg'))
plt.axis('off')
plt.subplot(222)
plt.imshow(mpimg.imread('business.jpg'))
plt.axis('off')
plt.subplot(223)
plt.imshow(mpimg.imread('workplace.jpg'))
plt.axis('off')
plt.subplot(224)
plt.imshow(mpimg.imread('children.jpg'))
plt.axis('off')
```

```python
fig = plt.figure(figsize=(12,10)) # Rozmiar to 1200x1000

for idx, img in enumerate (['office.jpg', 'business.jpg', 'workplace.jpg', 'children.jpg']):
    plt.subplot(220 + idx + 1)
    plt.imshow(mpimg.imread(img))
    plt.axis('off')
```
