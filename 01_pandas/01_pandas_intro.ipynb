### Podstawowe struktury danych: Series

```python
s = pd.Series(data=[3, 2, 4, 6])
s
```

```python
s = pd.Series(data=[3., 2, 4, 6], index=['a', 'b', 'c', 'd'], name='sample')
print(s)
```

```python
s = pd.Series(data=[3., np.nan, 4, 6], index=['a', 'b', 'c', 'd'], name='sample')
print(s)
```

```python
s = pd.Series(data=[True, False, False])
print(s)
```

```python
s = pd.Series(data=np.arange(10, 20), index=pd.date_range(start='20200101', periods=10))
print(s)
```

```python
price = pd.Series(data={'Apple': 200, 'CD Projekt': 60, 'Amazon': 1900})
print(price)
```

```python
price['CD Projekt']
```

```python
print(price[0])
```

```python
price.count() # Zliczy ile mamy elementów w obiekcie Series
```

```python
price.value_counts() # Zlicza ilość wystąpień danego elementu
```

```python
price.sum() # Sumuje wartość
```

```python
price.max() # max element
```

```python
price.min() # min element
```

```python
price.std() # Odchylenie standardowe
```

```python
price.describe() # Statystyki opisowe
```

```python
price = pd.Series(data={'Apple': 200, 'CD Projekt': 60, 'Amazon': 1900, 'KGHM': np.nan})
print(price)
```

```python
price.count()
```

```python
price.value_counts(dropna=False) # Ustawienie dropna=False pokaże ilość wystąpień braków
```

```python
price.nlargest(1) # Wyświetla największą wartość. W nawiasie możemy ustawić ile największych elementów ma pokazać (pomija element NaN)
```

```python
price.nsmallest(1) # Wyświetla najmniejszą wartość. W nawiasie możemy ustawić ile najmniejszych elementów ma pokazać (pomija element NaN)
```

```python
price.rank() # rankinkowanie elementów
```

```python
price.sort_values(ascending=False) # sortowanie wartości, domyślnie rosnąco, ascending=False malejąco
```

```python
price_pln = price.apply(lambda x: x * 4) # Mnoży wszystkie wartości razy 4
```

```python
price_pln
```

### DataFrame

```python
df = pd.DataFrame(data=[12, 12, 32])
print(df)
```

```python
df = pd.DataFrame(data=[12, 12, 32], index=['first', 'second', 'third'], columns=['col_1'])
print(df)
```

```python
df = pd.DataFrame(data={'WIG20': ['PKN ORLEN', 'PKO BP'],
                        'mWIG40': ['Amica', 'Playway']})
print(df)
```
