### Wykres liniowy

```python
_ = plt.plot([0, 1], [0, 3]) #[x,x],[y,y]
```

```python
_ = plt.plot([0, 1], [0, 3], label = 'line', color='green')
plt.xlabel('oś X')
plt.ylabel('oś Y')
plt.title('Wykres liniowy')
_ = plt.legend()
```

```python
_ = plt.plot([0, 1, 5], [0, 3 ,5], label='line', color='blue')
plt.legend()
plt.xlabel('oś X')
plt.ylabel('oś Y')
plt.title('Wykres liniowy')
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)
plt.plot(x, x_sin)
plt.plot(x, x_cos)
plt.legend(['sin', 'cos'])
```

### Tworzenie wielu wykresów

```python
plt.subplot(121) # co oznacza 121: 1 - Liczba wierszy wykresów, 2 - kolumny wykresów, 1- odnosimy sie do wykresu nr. 1(wykres niebieski)
plt.plot([0, 1, 2], [3, 2, 5], 'r--') # 'r--' - linia przerywana
plt.subplot(122) # tutaj 122 1- liczba wierszy wykresów, 2- liczba kolumn wykresów, 2 - odnosi się do wykresu nr. 2(wykres czerwony)
plt.plot([0, 1, 2], [3, 2, 5], color='red')
```

```python
fig = plt.figure(figsize=(10, 8)) # figsize - rozmiar wykresu w calach
plt.subplot(131)
plt.plot([0, 1, 2], [3, 2, 5], 'r--')
plt.subplot(132)
plt.plot([0, 1, 2], [3, 2, 5], color='red')
plt.subplot(133)
plt.plot([0, 1, 2], [3, 2, 5], 'bo') # b - odpowiada za kolor(b-blue), o-punktowy
fig.show() # w pycharm musi być, w notebookach nie musi tego być
```

### Wykres słupkowy

```python
plt.bar(x=[0, 1, 2, 3], height=[1.2, 3.2, 1.5, 2.8]) # pionowy
```

```python
plt.barh(y=['a', 'b', 'c', 'd'], width=[1.2, 3.2, 1.5, 2.8])  #poziomy
```

### Wykresy punktowe

```python
x1 = np.random.randn(300)
x2 = np.random.randn(300)
plt.scatter(x=x1, y=x2)
```

```python
x1 = np.random.randn(300)
x2 = np.random.randn(300)
x3 = np.random.randn(300) *50
plt.scatter(x=x1, y=x2, s=x3) # rozmiar kropek
```

```python
x1 = np.random.randn(300)
x2 = np.random.randn(300)
x3 = np.random.randn(300) *50
plt.scatter(x=x1, y=x2, s=x3, alpha=0.5) # alpha - przezroczystość
```
