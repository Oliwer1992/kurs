```python
import pandas as pd
import numpy as np
```

### Funkcja do pobierania danych gie≈Çdowych z stooq

```python
def fetch_finacial_data(company='AMZN'):
  import pandas_datareader.data as web
  return web.DataReader(name=company, data_source='stooq')
df = fetch_finacial_data('META')
df.head()
```

### Zapisywanie danych do pliku .csv

```python
df.to_csv('fb.csv')
```

```python
df_nov = df[(df.index.month == 5) & (df.index.year == 2022)]
```

```python
df_nov
```

```python
df_nov.to_csv('fb_nov.csv')
```

### Wczytywanie danych z pliku .csv

```python
new_df = pd.read_csv('fb_nov.csv', index_col=0)
```
